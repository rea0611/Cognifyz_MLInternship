```python
df = pd.read_csv("Dataset .csv")
print(df.columns)
print(df.head())
```

```python
df = df[['Restaurant Name', 'Cuisines', 'Price range', 'Aggregate rating']]

```

```python
df['Primary Cuisine'] = df['Cuisines'].apply(lambda x: x.split(',')[0].strip() if isinstance(x, str) else '')
```

```python
def recommend_restaurants(cuisine, price_range):
    # Filter by cuisine and price range
    filtered = df[(df['Primary Cuisine'].str.lower() == cuisine.lower()) & (df['Price range'] == price_range)]

    if not filtered.empty:
        print("‚úÖ Exact match found!")
    else:
        print("‚ö†Ô∏è No exact match. Trying same cuisine only...")
        filtered = df[df['Primary Cuisine'].str.lower() == cuisine.lower()]

        if filtered.empty:
            print("‚ö†Ô∏è No restaurants found for the selected cuisine. Showing all instead.")
            filtered = df

    return filtered.sort_values(by='Aggregate rating', ascending=False)
```

```python
user_cuisine = "Italian"
user_price = 2
```

```python
results = recommend_restaurants(user_cuisine, user_price)
```
